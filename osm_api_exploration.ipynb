🔹 Experiment 1: Extract Amenities

```python
import osmnx as ox

place = "Nancy, France"

tags_amenity = {"amenity": True}

gdf_amenity = ox.features_from_place(place, tags_amenity)

print("Number of amenities:", len(gdf_amenity))
gdf_amenity.head()
```

🔹 Experiment 2: Extract Tourism Locations

```python
tags_tourism = {"tourism": True}

gdf_tourism = ox.features_from_place(place, tags_tourism)

print("Number of tourism places:", len(gdf_tourism))
gdf_tourism.head()
```

🔹 Experiment 3: Extract Shops

```python
tags_shop = {"shop": True}

gdf_shop = ox.features_from_place(place, tags_shop)

print("Number of shops:", len(gdf_shop))
gdf_shop.head()
```
